## which genres have highest profit?

```python
movies['profit'] = movies['revenue'] - movies['budget']
```

```python
movies.head()
```

```python
# Removing extra columns
movies_genre= movies[['popularity','budget', 'revenue', 'original_title','runtime', 'genres', 'release_year', 'vote_average', 'vote_count', 'profit']]
```

```python
movies_genre.head()
```

```python
from pandas import Series, DataFrame
```

```python
split = movies_genre['genres'].str.split('|').apply(Series,1).stack()
split
```

```python
split.index = split.index.droplevel(-1)
```

```python
split
split.name = 'genres_split'
del movies_genre['genres']
movies_genre = movies_genre.join(split)
```

```python
movies_genre
```

## Which genres are the most common(number of movies made)?

```python
genres_count = pd.DataFrame(movies_genre.groupby('genres_split').original_title.nunique()).sort_values('original_title',ascending = False)
```

```python
genres_count['original_title'].plot.pie(title = 'Movies Per Genre in %', autopct= '%1.1f%%', figsize = (10,10))
```

```python
genres_count['original_title'].plot.barh(title = 'Movies Per Genre', color= 'DarkBlue', figsize = (10,9))
```

## Which genres have high avg budget and revenue?

```python
genres_avg = movies_genre.groupby('genres_split').mean(numeric_only = True)
pd.options.display.float_format = '{:.2f}'.format
genres_avg
```

```python
genres_avg.sort_values('budget', ascending = True, inplace = True)
```

```python
genres_avg[['budget', 'revenue']].plot.barh(title = 'Genres Hign Abg Budget', figsize = (9,9))
```

```python
genres_avg.sort_values('revenue', ascending = True, inplace = True)
genres_avg[['budget', 'revenue']].plot.barh(title = 'Genres Hign Abg Budget', figsize = (9,9))
```

## Which genres have highest number of movies with an voting avg >= 8 ?

```python
movies_genre.head()
```

```python
vote_fifty = movies_genre[(movies_genre['vote_count'] > 50) & (movies_genre['vote_average'] > 8)]
```

## The effect of vote avg on budget and revenue

```python
movies.drop_duplicates(inplace = True)
movies['profit'] = movies['revenue'] - movies['budget']
movies_genre= movies[['popularity','budget', 'revenue', 'original_title','runtime', 'genres', 'release_year', 'vote_average', 'vote_count', 'profit']]
movies_genre.head()
```

```python

movie_counted = movies_genre[movies_genre['vote_count'] > 50]
movie_counted.corr(method ='spearman', numeric_only = True)
```

```python
import seaborn as sns

sns.regplot(x= 'vote_average', y= 'profit', data = movie_counted, line_kws = {'color': 'red'})
```

## Effect of popularity on profit and revenue

```python
movie_counted.corr(method ='spearman', numeric_only = True)
```

```python
import matplotlib.pyplot as plt
sns.regplot(x= 'popularity', y= 'profit', data = movie_counted, line_kws = {'color': 'red'})
plt.title('Effect of popularity on profit')
```

```python
import matplotlib.pyplot as plt
sns.regplot(x= 'popularity', y= 'revenue', data = movie_counted, line_kws = {'color': 'red'})
plt.title('Effect of popularity on revenue')
```

## Effect of budget on profit

```python
movie_counted.corr(method ='spearman', numeric_only = True)
```

```python
sns.regplot(x= 'budget', y= 'profit', data = movie_counted, line_kws = {'color': 'red'})
plt.title('Effect of budget on profit')
```

## Highly budgeted movie has high popularity

```python
sns.regplot(x= 'budget', y= 'popularity', data = movie_counted, line_kws = {'color': 'red'})
plt.title('Effect of budget on popularity')
```

```python
timegenre= pd.DataFrame(movies_year.groupby(['release_year', 'genres_split'])['profit'].mean())
```

```python
timegenre
```

```python
final_genre = pd.pivot_table(timegenre, values= 'profit', index= ['genres_split'], columns = ['release_year'])
```

```python
sns.set(rc= {'figure.figsize':(15,10)})
sns.heatmap(final_genre, cmap = 'YlGnBu', linewidths = 0.5)
plt.title('Genres By Profit By Year')
plt.show()
```
